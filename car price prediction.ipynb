```python
df2=df2.drop(columns=['carlength'])
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the features to be normalized into a separate DataFrame
features_to_normalize = df2[['horsepower', 'enginesize', 'carwidth','curbweight']]

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the features
scaler.fit(features_to_normalize)

# Transform the features using the scaler
normalized_features = scaler.transform(features_to_normalize)

# Replace the original features in the DataFrame with the normalized ones
df2[['horsepower', 'enginesize', 'carwidth','curbweight']] = normalized_features
```

```python
df2.head()
```

```python
data=df2.values
x=data[:,0:4]

y=data[:,4]
```

```python
#train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
lr.predict([[0.174483,	0.074449,	-0.844782,	-0.014566]])
```

```python
from sklearn.metrics import r2_score
ac=r2_score(ytest, ypred)
ac
```
